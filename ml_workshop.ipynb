# Basic Computing in NumPy
NumPy is a numerical processing module for python, built on top of hardcore C and Fortran libraries.  This means it's actually pretty fast yet still gives you easy python programming.  Let's explore some of the nice things of NumPy and how it's different from standard Python.

## NumPy arrays vs Python Lists
A standard Python allows you to put arbitrary objects into an ordered list, e.g.

```python
L = [3, 'abc', math.sin]
```

Note that you can put anything into a list, a number, a string, a function; pick your favorite object.  And you can put different types into a single list.  It's all very convenient...and very inefficient.  Every time you step through the list, Python has to do math behind the scenes to pull out the correct element.  But imagine if everything in your list was the same type and everything took up the same size.  You could quickly jump to any location.  That's basically what a NumPy array does; it's a block of memory inside your computer.  You lose some of that generality in order, but gain a ton of speed and storage savings.

````python
import numpy as np

# Arrays come in all shapes and values...let's start with zeros
z = np.zeros(10)
print(z)

# Or fill it up with ones
o = np.ones(20)
print(o)

# Arrays can be multidimensional, x, y, z, ....as many as you need, and you can make them from proper lists
n = np.array([[1,2],[3,4]])
print(n)

# Numpy even support random generation of matrices
r = np.random.randint(2,size=[2,2])
print(r)
````

## Slice and Dice
Python lists (and other "iterables") let you quickly select part of the object (i.e. the first 10 entries).  NumPy arrays crank this up to eleven, giving you full control of accessing arbitrary submatrices quickly and conveniently.

````python
# Here's our list
a = [i for i in range(20)]
# Skip the first 5 elements
print(a[5:])
# Stop at the 10th element (exclusive)
print(a[:10])
# Stop 2 back from the end
print(a[:-2])
# Take every other element
print(a[::2])
# Generally a[start:stop:skip]

# Now get an array, nicely numbered
b = np.arange(100).reshape(10,10)
print(b)
# Typically access an element with comma-separated indices
print(b[1,2])
# Grab an entire row, or column, or whatever
print(b[0,:]) # colon is shorthand for everything
print(b[:,0])
# Access only certain elements
print(b[[1,2,3,4],[6,7,8,9]])
# Use truth values to pull out certain elements into a new array
print(b[b%2 == 1])
````

## Linear Algebra
Wait!  Don't run away!  NumPy can do the linear algebra for you!  It uses some non-obvious terminology, but it can handle typical matrix operations.

````python
B = np.random.randint(10,size=[2,2])
print(B)
C = np.random.randint(10,size=[2,2])
print(C)
# Element-wise, 
print(B*C)
# Actual matrix multiplication is a little funny, I wish it was just np.matmul
print(np.dot(B,C))
# Transpose
print(B.T)
# inversion (or pseudo-inversion if necessary)
print(np.linalg.pinv(B))
````

# scikit-learn
[scikit-learn](scikit-learn.org/stable) is a Python library that efficiently implements a ton of common statistical models.  Linear and logistic regression, k-means clustering, you name it.  Let's learn some basics.

````python
import sklearn
from sklearn import datasets, linear_model
````

````python
# Standard Linear Regression, aka Ordinary Least Squares
## Fabricate some data
Y = np.arange(10) + np.random.randn(10)
X = np.arange(10)

# Always good to take a look if you can
plt.scatter(X,Y)

lr = linear_model.LinearRegression()

# scikit-learn is very general, so you might have to put your data into a slightly different format
lr.fit(X.reshape(10,1),Y.reshape(10,1))
plt.plot(X,lr.predict(X.reshape(10,1)))
````

````python
# Classification with Logistic Regression, let's use scikit-learn's built-in datasets

# load the data
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
Y = iris.target

# This does the actual fitting
clf = linear_model.LogisticRegression()
clf.fit(X,Y)

# This just makes a pretty plot of what regions are what
h = 0.02 # how dense the mesh of points are
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
# foo.ravel is awesome for arrays; it unrolls all the dimensions, giving you a 1-D array
# This makes for easy predicting
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', s=100, lw = 2, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
````

# Machine Learning and Neural Nets
TensorFlow is Google's new library for graph-based computing.  That just means that when you run most of the commands, they just tell TF how to do a computation, but don't actually do it yet.  This is actually nice for applications like neural networks, where you define the shape of net, and then train it.  TF includes a number of tools to help make neural nets; we're just going to scratch the surface with a simple network to classify some handwritten digits.

````python
import skflow
from sklearn import datasets, cross_validation, metrics
````

````python
# Whip up a deep neural net to classify those irises, say 3 layers of interior neurons
skf = skflow.TensorFlowDNNClassifier(hidden_units=[10, 20,10], n_classes=3, steps=200)

skf.fit(X,Y)
````

````python
# Let's replot and see how this did
Z = skf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(1)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', s=100, lw = 2, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
````
